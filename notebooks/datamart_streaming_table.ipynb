```sql
USE CATALOG demo;
USE SCHEMA bronze;
DROP TABLE IF EXISTS customer_ingest;

CREATE OR REFRESH STREAMING TABLE customer_ingest COMMENT "Ingest Data in AS IS format" 
AS 
SELECT * 
FROM STREAM read_files(
  "dbfs:/databricks-datasets/tpch/data-001/customer", 
  format=>"csv",
  schema => "c_custkey INT, c_name STRING, c_address STRING, c_nationkey INT, c_phone STRING, c_acctbal DECIMAL(12,2), c_mktsegment STRING, c_comment STRING", 
  delimiter=> "|"
  );

select * from customer_ingest limit 10;
```

```sql
USE CATALOG demo;

USE SCHEMA bronze;

DROP TABLE IF EXISTS order_ingest;

CREATE OR REFRESH STREAMING TABLE order_ingest COMMENT "Ingest Data in AS IS format" 
AS 
SELECT * 
FROM STREAM read_files(
  "dbfs:/databricks-datasets/tpch/data-001/orders",
  format => "csv",
  schema => "o_orderkey int, o_custkey int, o_orderstatus string, o_totalprice decimal(15,2), o_orderdate date, o_orderpriority string, o_clerk string, o_shippriority int, o_comment string",
  delimiter => "|"
  );

select * from order_ingest limit 10;
```

```sql
USE CATALOG demo;

USE SCHEMA bronze;

DROP TABLE IF EXISTS lineitem_ingest;

CREATE OR REFRESH STREAMING TABLE lineitem_ingest COMMENT "Ingest Data in AS IS format"
AS 
SELECT * 
FROM STREAM read_files(
  "dbfs:/databricks-datasets/tpch/data-001/lineitem", 
  format=>"csv",
  schema => "l_orderkey int, l_partkey int, l_suppkey int, l_linenumber int, l_quantity decimal(15,2), l_extendedprice decimal(15,2), l_discount decimal(15,2), l_tax decimal(15,2), l_returnflag string, l_linestatus string, l_shipdate date, l_commitdate date, l_receiptdate date, l_shipinstruct string, l_shipmode string, l_comment string", 
  delimiter=> "|"
  );

select * from lineitem_ingest limit 10;
```

```sql
USE CATALOG demo;

USE SCHEMA bronze;

DROP MATERIALIZED VIEW IF EXISTS customers_mv;

CREATE MATERIALIZED VIEW customers_mv
AS 
SELECT c_custkey, c_name, c_nationkey, c_acctbal
FROM demo.bronze.customer_ingest
WHERE c_custkey IS NOT NULL
AND c_nationkey != 21;

select * from customers_mv limit 10;
```

```sql
USE CATALOG demo;

USE SCHEMA bronze;

DROP MATERIALIZED VIEW IF EXISTS demo.bronze.orders_mv;

CREATE MATERIALIZED VIEW demo.bronze.orders_mv
AS 
SELECT * 
FROM demo.bronze.order_ingest
WHERE o_orderkey is not null
AND o_orderdate > '1990-01-01';

select * from demo.bronze.orders_mv limit 10;
```

```sql
USE CATALOG demo;

USE SCHEMA bronze;

DROP MATERIALIZED VIEW IF EXISTS demo.bronze.liteitem_mv;

CREATE MATERIALIZED VIEW demo.bronze.liteitem_mv
AS 
SELECT l_orderkey, l_linenumber, l_quantity, l_extendedprice, l_discount, l_returnflag, l_linestatus
FROM demo.bronze.lineitem_ingest
WHERE l_orderkey IS NOT NULL
AND l_shipdate > '1990-01-01';

select * from demo.bronze.liteitem_mv limit 10;
```

```sql
USE CATALOG demo;
USE SCHEMA silver;

-- customer_silver
drop table if exists customers_silver;
create table customers_silver
as
select * from demo.bronze.customers_mv;
alter table customers_silver alter column c_custkey set not null;
alter table customers_silver add constraint customer_pk primary key(c_custkey);

select * from customers_silver limit 10;
```

```sql


-- orders_silver
drop table if exists orders_silver;
create table orders_silver
as
select * from demo.bronze.orders_mv
where o_orderkey is not null
and o_orderdate > '1990-01-01';
alter table orders_silver alter column o_orderkey set not null;
alter table orders_silver add constraint order_silver_pk primary key(o_orderkey);
alter table orders_silver alter column o_custkey set not null;
alter table orders_silver add constraint order_fk foreign key(o_custkey) references customers_silver(c_custkey);

select * from orders_silver limit 10;


```

```sql
-- lineitem_silver
drop table if exists lineitem_silver;
create table lineitem_silver
as
select * from demo.bronze.liteitem_mv
where l_orderkey is not null;
alter table lineitem_silver alter column l_linenumber set not null;
alter table lineitem_silver add constraint lineitem_silver_pk primary key(l_linenumber);
alter table lineitem_silver alter column l_orderkey set not null;
alter table lineitem_silver add constraint lineitem_fk foreign key(l_orderkey) references orders_silver(o_orderkey);

select * from lineitem_silver limit 10;
```

```sql
USE CATALOG demo;
USE SCHEMA gold;

create materialized view order_status as
select 
  o_orderstatus as status, 
  count(*) as num_of_orders
from demo.silver.orders_silver
group by o_orderstatus;

select * from order_status limit 10;

```

```sql
USE CATALOG demo;
USE SCHEMA gold;

create materialized view total_order_10000 as
SELECT o_orderdate, SUM(o_totalprice) AS total_order
FROM demo.silver.orders_silver
GROUP BY o_orderdate
HAVING SUM(o_totalprice) > 10000;

select * from total_order_10000 limit 10;
```
